###Standardize text fields to prevent filter errors

```python

df["product_name"] = df["product_name"].str.strip().str.lower()
df["payment_method"] = df["payment_method"].str.strip().str.lower()
df["store"] = df["store"].str.strip().str.lower()
df["customer_type"] = df["customer_type"].str.strip().str.lower()
```

###CHECK FOR UNIQUE VALUES

```python

print(df['product_name'].unique())
```

###1. Total number of apples purchased using cash

```python

apples_cash_quantity = df[
    (df["product_name"] == "apple") &
    (df["payment_method"] == "cash")
]["quantity"].sum()
print(f"Total number of apples purchased using cash: {apples_cash_quantity}")

```

###2. Total cash amount spent on apples

```python

apples_cash_amount = df[
    (df["product_name"] == "apple") &
    (df["payment_method"] == "cash")
]["total_amount"].sum()

print(f"Total cash amount spent on apples: {apples_cash_amount}")
```

###3. Total amount spent by non-member customers at Bakershire

```python

non_member_bakershire_spend = df[
    (df["customer_type"] == "non-member") &
    (df["store"] == "bakershire")
]["total_amount"].sum()

print(f"Total amount spent by non-member customers at Bakershire: {non_member_bakershire_spend}")
```
